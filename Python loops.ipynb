```python
import pandas as pd


print("Numbers from 1 to 20:")
for i in range(1, 21):
    print(i)

# Print only even numbers from 1 to 20
print("Even numbers from 1 to 20:")
for i in range(1, 21):
    if i % 2 == 0:
        print(i)
```

```python
num = 10

while num >= 1:
    print(num)
    num -= 1
```

```python
nums = [3, 10, 15, 8, 21, 4]

for n in nums:
    if n > 10:
        print(n)
```
