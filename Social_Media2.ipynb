```python
import pandas as pd
df=pd.read_csv("Movie_review.csv")
df
```

```python
print(df.head(10))
print(df.info())
print(df.describe())
```

```python
print(df.duplicated())
```

```python
print(df.duplicated().sum())
```

```python
df = df.drop_duplicates()
```

```python
print(df.isnull())
```

```python
print(df.isnull().sum())
```

```python
df_cleaned = df.dropna()
```

```python
df['timestamp'] = pd.to_datetime(df['timestamp'], errors='coerce')
df.info()
```

```python
print(df.dtypes)
```

```python
df = df.apply(lambda col: col.str.strip() if col.dtype == "object" else col)
df
```

```python
df.columns = df.columns.str.title()
df
```

```python
df['Sentiment']=df['Sentiment'].str.capitalize()
df
```

```python
df['Movie_Name']=df['Movie_Name'].str.capitalize()
df['Movie_Review']=df['Movie_Review'].str.capitalize()
df['Platform']=df['Platform'].str.capitalize()
df['Duration']=df['Duration'].str.capitalize()
df
```

```python
df["Duration"] = df["Duration"].str.replace("min", " Min", regex=False)
df
```

```python
df['Rating']=df['Rating'].fillna(df['Rating'].mean())
df
```

```python
df['Likes']=df['Likes'].fillna(df['Likes'].median())
df
```

```python
df["Likes"] = df["Likes"].replace(-25, 25)
df
```

```python
df.sort_values(by='Rating',ascending=False)
```

```python
import numpy as np
df['User_Id'] = pd.to_numeric(df['User_Id'], errors='coerce')

#count missing ids
n_missing = df['User_Id'].isna().sum()

#generate random values between 1 and 999
random_ids = np.random.randint(1, 1000, size=n_missing)

#fill the missing values
df.loc[df['User_Id'].isna(), 'User_Id'] = random_ids
df['User_Id'] = df['User_Id'].astype(int)

# Convert User_Id to string and add the desired prefix and format
df['User_Id'] = 'User_id ' + df['User_Id'].astype(str)
df
```

```python
df.rename(columns={'Review_Text': 'Movie_Review'}, inplace=True)
df
```

```python
df["weekday"] = df["Timestamp"].dt.day_name()
df
```

```python
df["Month_Name"] = df["Timestamp"].dt.month_name()
df
```

```python
df["Rating_Review"] = df["Rating"].apply(
    lambda x: "Flop" if x < 4 else "Average" if x <= 8 else "Hit"
)
df
```

```python
import matplotlib.pyplot as plt

stages = ["Views", "Likes", "Reviews", "Positive Reviews"]
values = [1000, 600, 300, 150]

# Reverse for top-down funnel
stages = stages[::-1]
values = values[::-1]

left = [-v for v in values]
right = values

plt.figure(figsize=(7,5))
plt.barh(stages, left)
plt.barh(stages, right)

for i, v in enumerate(values):
    plt.text(0, i, str(v), ha='center', va='center')

plt.title("Funnel Chart of Movie Engagement")
plt.xticks([])
plt.show()
```

```python
import matplotlib.pyplot as plt

# Funnel data
stages = ["Lead", "Opportunity", "Qualified", "Sold", "Retained"]
values = [7745, 5468, 4238, 2025, 1520]

# Reverse for correct funnel display
stages = stages[::-1]
values = values[::-1]

# Create figure
plt.figure(figsize=(8, 6))

# Draw horizontal bars
plt.barh(stages, values)

# Add value labels with percentage
total = values[-1]
for i, v in enumerate(values):
    percent = (v / total) * 100
    plt.text(v + 50, i, f"{v} | {percent:.1f}%", va="center")

# Labels and title
plt.xlabel("Value")
plt.title("Sales Funnel")

plt.show()
```

```python
import matplotlib.pyplot as plt

# Funnel data
stages = ["Lead", "Opportunity", "Qualified", "Sold", "Retained"]
values = [7745, 5468, 4238, 2025, 1520]

# Reverse for top-to-bottom display
stages = stages[::-1]
values = values[::-1]

# Create left and right values
left = [-v for v in values]
right = values

plt.figure(figsize=(8, 6))

# Plot both sides
plt.barh(stages, left, color="#5DADE2")
plt.barh(stages, right, color="#2E86C1")

# Add labels
max_val = max(values)
for i, v in enumerate(values):
    percent = (v / max_val) * 100
    plt.text(0, i, f"{v} | {percent:.1f}%", ha="center", va="center", color="black")

# Remove x-axis ticks
plt.xticks([])
plt.xlabel("Value")
plt.title("Sales Funnel (Both Sides)")

plt.show()
```
